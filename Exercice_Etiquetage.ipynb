```python
#Jeu d'étiquettage Treetagger sur une partie du corpus: ##D'après le fichier txt que j'ai construit Mini_Corpus_Annoté 
import re

with open("mini_corpus.tagxt", encoding="utf-8") as rp:
    corpus = []
    for p in rp:
        phrase = []
     
        for t in p.rstrip().split("\t"):
            elts = re.split("===", t)
         
            if len(elts) >= 2:
                mot = elts[0]
                etiquette = elts[-1] 
                phrase.append((mot, etiquette))
        corpus.append(phrase)
```
